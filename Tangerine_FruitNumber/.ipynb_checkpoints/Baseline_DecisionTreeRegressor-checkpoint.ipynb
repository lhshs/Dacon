## Optuna Tuning

```python
from sklearn.model_selection import train_test_split
```

```python
import optuna
from optuna import Trial, visualization
from optuna.samplers import TPESampler
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop(['ID','착과량(int)'], axis=1), 
                                                    train['착과량(int)'], test_size=0.2, random_state = 1042)
```

```python
sampler = TPESampler(seed=20) # 젤 많이 쓰이는 sampler, DL에서 optimizer 느낌

def objective(trial):
    dtrain = Dataset(X_train, label = y_train)
    dtest = Dataset(X_test, label = y_test)

    params = {
        'objective': 'regression', # 회귀
        'verbose': -1,
        'metric': 'rmse', 
        'boosting_type': 'gbdt',
        'max_depth': trial.suggest_int('max_depth', 5, 50),        # suggest_int 정수로 이동하면서 학습
        'learning_rate': trial.suggest_loguniform("learning_rate", 1e-8, 1e-2), # suggest_loguniform 함수를 써 줘야 소수점을 따라 학습
        'n_estimators': trial.suggest_int('n_estimators', 1, 3000),
        'min_child_samples': trial.suggest_int('min_child_samples', 1, 50),
        'subsample': trial.suggest_loguniform('subsample', 0.2, 1),
    }
    
    model = LGBMRegressor(**params) # ** = 가변적이라는 뜻
    lgb_model = model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=0, early_stopping_rounds=25)
    mse = mean_squared_error(y_test, lgb_model.predict(X_test))
    return mse

study_lgb = optuna.create_study(direction='minimize', sampler=sampler)
study_lgb.optimize(objective, n_trials=50)

```

```python
op_params1 = {'max_depth': 15, 'learning_rate': 0.0016098380341155907, 'n_estimators': 2982, 
              'min_child_samples': 50, 'subsample': 0.21700326775522788}
```

```python
lgb_model = LGBMRegressor(**op_params1)
lgb_model.fit(X_train, y_train)

# mean_squared_error(lgb_pred, y_test.values)
```
